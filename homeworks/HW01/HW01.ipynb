```python
grades = np.array([6, 7, 8, 9, 10])

print("Оценки:", grades)
print("Средний балл:", grades.mean())

grades_min = grades.min()
grades_max = grades.max()
norm_grades = (grades - grades_min) / (grades_max - grades_min)

print("Нормализованные оценки:", norm_grades)
```
